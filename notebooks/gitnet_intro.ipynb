# Getting Logs with Gitnet

```python
import os
import gitnet as gx
```

```python
# This will navigate to the data folder within the gitnet notebooks. This code will only run properly once.
# You can test it, making sure that you are in the data folder, by printing CurrentDir.

CurrentDir = os.getcwd() # get the current working directory and store it for later
CurrentDir = os.chdir(CurrentDir + '/data/')

print(CurrentDir)
```

```python
# Enter the path to any directory that contains a .git folder. The default will use the gitnet repo itself.
mylog = gx.get_log(CurrentDir)
```

```python
# Describe is the most basic feature in gitnet. It returns a summary of the log object.

mylog.describe()
```

```python
# Attributes returns a list of the keys that make up each record in the log object.

mylog.attributes()
```

```python
# Filter can be used to create a filtered log object, which contains only records with particular values.

myfilteredlog = mylog.filter('email', 'has', '@uwaterloo')
```

```python
# One of the most useful features is exporting a log object to a TSV file. 

mylog.tsv(fname = 'test_mylog.tsv')
```

```python
# Log objects can also be exported to Pandas dataframes.

mylog.df()
```

```python
# Vectors are useful for extracting the plain text output from a specific field in the git log, for example, messages for qualitative analysis.

mylog.vector('message')
```

```python
# Browse moves through each entry one at a time.

mylog.browse()
```
